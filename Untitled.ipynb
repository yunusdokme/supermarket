```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=train_data.drop(["Row ID","Order ID","Customer ID","Product ID"],axis=1)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df[df["Postal Code"].isnull()]
```

```python
df["Postal Code"] = df["Postal Code"].fillna(5402)
```

```python
df[df["Postal Code"].isnull()]
```

```python
df[df["City"]=='Burlington']
```

```python
df["Month"]= df["Order Date"].str[3:5]
df
```

```python
df["Month"]=df["Month"].astype("int32")
df
```

```python
d={1:"Ocak",2:"Şubat",3:"Mart",4:"Nisan",5:"Mayıs",6:"Haziran",7:"Temmuz",8:"Ağustos",9:"Eylül",10:"Ekim",11:"Kasım",12:"Aralık"}
d
```

```python
df["Month"]=[d[i] for i in df["Month"]]
df
```

```python
df["Order Date"]=pd.to_datetime(df["Order Date"],dayfirst=True)
```

```python
df
```

```python
df["Ship Date"]=pd.to_datetime(df["Ship Date"],dayfirst=True)
df
```

```python
df["Year"]=df["Order Date"].dt.year
df
```

```python
df["Year"].value_counts()
```

```python
df["Month"].value_counts()
```

```python
months_df = df.groupby("Month").sum(numeric_only=True)
```

```python
months_df
```

```python
months_df = months_df.sort_values("Sales")
months_df
```

```python
months_df.reset_index(inplace=True)
months_df
```

```python
plt.bar(months_df["Month"],months_df["Sales"])
plt.xlabel("Aylar")
plt.ylabel("Satış miktarı")
plt.xticks(fontsize=14,rotation=90)
plt.yticks(fontsize=14)
plt.show()
```

```python
year_df=df.groupby("Year").sum(numeric_only=True)
year_df=year_df["Sales"]
year_df
```

```python
year_df = df.groupby("Year")[["Sales"]].sum()
year_df.reset_index(inplace=True)
```

```python
plt.bar(year_df["Year"],year_df["Sales"])
plt.xlabel("Yıl")
plt.ylabel("Satış miktarı")
plt.xticks(fontsize=5,rotation=90)
plt.yticks(fontsize=4)
plt.show()
```

```python
Customer=df.groupby("Customer Name").sum(numeric_only=True).sort_values("Sales",ascending=False)
Customer.drop(["Postal Code","Year"],axis=1,inplace=True)
Customer
```

```python
Customer.reset_index(inplace=True)
Customer
```

```python
Customers=Customer.head(20)
```

```python
plt.bar(Customers["Customer Name"],Customers["Sales"])
plt.xlabel("Müşteri")
plt.ylabel("Satış miktarı")
plt.xticks(fontsize=14,rotation=90)
plt.yticks(fontsize=14)
plt.show()
```

```python
category=df.groupby("Category").sum(numeric_only=True)
category
```

```python
category=df.groupby("Category").sum(numeric_only=True)
category.drop(["Postal Code","Year"],axis=1,inplace=True)
category
```
